# Generate prompts

```typescript
import { PromptGenerator } from '../mod.ts';

const generator = new PromptGenerator(
    {
        defaultCount: 30,
        language: 'English',
        topics: [
            { topic: 'AI visibility', subtopics: ['ranking factors', 'prompt coverage'] },
            'brand monitoring'
        ],
        personas: [
            { name: 'SEO lead', description: 'Owns search visibility strategy at a SaaS company' },
            'Founder'
        ],
        intents: ['informational', 'comparison', 'transactional']
    },
    { model: 'gpt-4.1' }
);

const result = await generator.invoke('Generate prompts for a B2B AI visibility tool.');
console.log(result.parsed);
```
